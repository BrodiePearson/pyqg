```python
# imports and stuff
import twolayer_qg
import particles
import time
%pylab inline
```

# Equations

$$ \partial{q}{t} + J(\psi, q) = 0 $$

```python
# setup model
year = 24*60*60*360.
m = twolayer_qg.QGModel(tavestart=0,  dt=8000, tmax=4.01*year)
```

```python
# set up lagrangian particles

#Npart = 100 # number of particles
#x0 = rand(Npart)*m.L
#y0 = rand(Npart)*m.W

x0,y0 = meshgrid(m.x[0,::1], m.y[::1,0])
x0 = x0.ravel()
y0 = y0.ravel()
Npart = x0.size

lpa = particles.LagrangianParticleArray2D(x0, y0, 
        periodic_in_x=True, periodic_in_y=True,
        xmin=0, xmax=m.L, ymin=0, ymax=m.W)

# number of particles to save
Nhold = 3
particle_history = zeros((Nhold,2,Npart))
```

```python
%%prun

# hack to reset the timer
if m.t > 4*year:
    m.t = 4*year

started_advecting = False
tprev = time.time()
n=0

# set up extended grid for lagrangian particles
x = np.hstack([m.x[0,0]-m.dx, m.x[0,:], m.x[0,-1]+m.dx])
y = np.hstack([m.y[0,0]-m.dy, m.y[:,0], m.y[-1,0]+m.dy])
for snapshot in m.run_with_snapshots(
        tsnapstart=4*year, tsnapint=m.dt):

    if n==0:
        qi = m.q1.copy()
        ui = m.u1.copy()
        vi = m.v1.copy() 
    
    # set up velocities for lagrangian advection
    # need a view with wrapped values
    u = m.u1[r_[-1,0:m.ny,0]][:,r_[-1,0:m.nx,0]]
    v = m.v1[r_[-1,0:m.ny,0]][:,r_[-1,0:m.nx,0]]

    if started_advecting:
        lpa.runge_kutta_step(
            x, y, uprev, vprev, u, v, m.dt)
        particle_history = np.roll(
            particle_history, 1, axis=0
        )
        particle_history[0,0,:] = lpa.x
        particle_history[0,1,:] = lpa.y
        n+=1
    
    uprev = u.copy()
    vprev = v.copy()
    started_advecting = True
    if n>10:
        break

# now the model is done    

```

```python
# daily plots
clf()
print n
fig=figure(figsize=(12,12))
ax = subplot(111)
hold(True)
contourf(m.x, m.y,
    m.q1 + m.beta1*m.y, linspace(0, m.beta1*m.W,20),extend='both', cmap='RdBu_r')
quiver(m.x,m.y,m.u1,m.v1, scale=10)
plot(particle_history[-1,0,:].ravel(),
         particle_history[-1,1,:].ravel(), 'o', color='0.5') 
plot(particle_history[0,0,:].ravel(),
         particle_history[0,1,:].ravel(), 'ko')
#plot(x0.ravel(), y0.ravel(), 'ro')
ax.set_xticks([])
ax.set_yticks([])
```

```python
# calculate CG Tensor
C = zeros((m.ny,m.nx, 2, 2))

x0.shape = (Npart,)
y0.shape = (Npart,)

# final positions
x1 = particle_history[0,0,:]
y1 = particle_history[0,1,:]
# x displacement
dx = x1-x0
dy = y1-y0
# roll displacements across the borders
dx[ dx > m.L/2 ] -= m.L
dx[ dx < -m.L/2 ] += m.L
dy[ dy > m.W/2 ] -= m.W
dy[ dy < -m.W/2 ] += m.W

# make them 2D
dx.shape = m.x.shape
dy.shape = m.y.shape

# particles were initialized on the grid
DX, DY = m.dx, m.dy
# second-order centered differnce for gradient
C[:,:,0,0] = (roll(dx, -1, axis=1) - roll(dx, 1, axis=1))/(2*DX)
C[:,:,0,1] = (roll(dx, -1, axis=0) - roll(dx, 1, axis=0))/(2*DY)
C[:,:,1,0] = (roll(dy, -1, axis=1) - roll(dy, 1, axis=1))/(2*DX)
C[:,:,1,1] = (roll(dy, -1, axis=0) - roll(dy, 1, axis=0))/(2*DY)

# calculate trace and determinant
T = C[:,:,0,0] + C[:,:,1,1]
D = C[:,:,0,0]*C[:,:,1,1] - C[:,:,0,1]*C[:,:,1,0]

eval1 = 0.5*T - np.sqrt(0.25*T**2 - D + 0j)
eval2 = 0.5*T + np.sqrt(0.25*T**2 - D + 0j)

#evals = zeros((m.ny, m.nx, 2), dtype=complex64)
#evecs = zeros_like(C, dtype=complex64)
#for j in range(m.ny):
#    for i in range(m.nx):
#        evals[j,i], evecs[j,i] = eig(C[j,i])
#assert all

figure(figsize=(14,8))
subplot(121)
imshow(D); colorbar(orientation='horizontal')
subplot(122)
imshow(T); colorbar(orientation='horizontal')

print eval1
    
```
